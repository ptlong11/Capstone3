There are a lot of acronyms used in this data set. like SRS and ORtg_x. I will create a dictionary here to refer to, and will copy this for the EDA, Pre-Process, and Modeling portions of this project.

The Sport Reference data has 9840 entries, with 40 columns.

SRS : Team SRS (simple rating system) is a team evaluation rating that takes into account average point differential and strength of schedule.

'Pace' : the number of possessions a team uses per game

'Rel_Pace' : Real Pace

'ORtg_x' : Offensive Rating 

'Rel_ORtg' : Real Offensive Rating 

'DRtg_x' : Defensive Rating

'Rel_DRtg' : Real Defensive Rating

'Top WS' : Win Share

'Tm' : Team

'G' : Games played

'GS' : Game started

'MP' : Minutes Played

'PER' : The player efficiency rating (PER) is a rating of a player's per-minute productivity.

'3PAr' : 3-Point Field Goal Attempts

'FTr' : What is ftr NBA?
Free Throw Rate or FTR is the ratio of Free Throw Attempts to Field Goal Attempts

'ORB%' : Offensive Rebounds

'DRB%' : Defensive Rebound Percentage

'TRB%' : Total Rebounds

'AST%' : Assist percentage

'STL%' : Steals

'BLK%' : Blocks

'TOV%' : Turnovers

'USG%' Usage Rate or Usg is an estimate of the percentage of a team's offensive possessions used by an individual player during his time on the floor.

'OWS': Offensive Win Shares is a player statistic which attempts to divvy up credit for team success to the individuals on the team.

'DWS' : Defensive Win shares

'WS' : winshares

'WS/48' : Wineshares per 48 minutes

'OBPM' : Offensive Box Plus/Minus

'DBPM' : Defensive Box Plus/Minues

'BPM' : Box plus-minus is based on box score for evaluating basketball players' quality and contribution to the team from play-by-play regression

'VORP' : Value Over Replacement Player
